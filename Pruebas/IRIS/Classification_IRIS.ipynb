```python
import os
import lazypredict
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

np.random.seed(42)
```

<a id="section11"></a>
# <font color="#004D7F" size=5> 1.1. Read the dataset</font>

```python
#Read CSV
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['class'] = iris.target

df
```

```python
df_x = df.drop('class', axis = 1)
df_y = df['class']
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_y = label_encoder.fit_transform(df_y)

labels = label_encoder.classes_

for label, integer_value in zip(labels, range(len(labels))):
    print(f"Label: {label} -> Integer Value: {integer_value}")
```

<a id="section11"></a>
# <font color="#004D7F" size=5> 1.2. Preparing experimentation</font>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.30, random_state = 42,stratify=df_y)
#X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size = 0.30, random_state = 42,stratify=y_val)
```

```python
df_train = pd.concat([X_train, pd.DataFrame({'class': y_train})], axis = 1)
df_test = pd.concat([X_test, pd.DataFrame({'class': y_test})], axis = 1)
#df_val = pd.concat([X_val, pd.DataFrame({'class': y_val})], axis = 1)
```

```python
print(df_train['class'].value_counts())
print(' ')
print(' ')
print(df_test['class'].value_counts())
print(' ')
print(' ')
#print(df_val['class'].value_counts())
```

# <font color="#004D7F" size=5> 2.0. Lazy Predict</font>

```python
from lazypredict.Supervised import LazyClassifier

clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None, random_state=42)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)

models
```
